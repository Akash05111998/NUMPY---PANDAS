```python
# 14. Find Records for June 2014

ne['Order_Date']=pd.to_datetime(ne['Order_Date'])
# june_2014=ne[(ne['Order_Date'] >= '2014-06-01') & (ne['Order_Date'] <= '2014-06-30')]
june_2014=ne[(ne['Order_Date'].dt.month==6) & (ne['Order_Date'].dt.year==2014)]

june_2014
```

```python
# 15. Calculate month wise average of profit. 

ne['Order_Date']=pd.to_datetime(ne['Order_Date'])  
monthly_avg_profit=ne.groupby(ne['Order_Date'].dt.month)['Profit'].mean()
monthly_avg_profit
```

```python
# 16. Create the new data frame for California 

california_data = ne[ne['State'] == 'California']
california_data
```

```python
# 17. Find sum of sales  

sum_of_sales = ne['Sales'].sum()
sum_of_sales
```

```python
#18 Find state wise sum of sales.

aaaa=ne.groupby('State')['Sales'].sum()
aaaa
```

```python
#19. Find month wise sum  of sales 

bbb=ne.groupby([ne['Order_Date'].dt.year, ne['Order_Date'].dt.month])['Sales'].sum()
bbb
```

```python
#20. Find country wise and city wise average profit

ccc=ne.groupby(['Country','City'])['Profit'].mean()
ccc
```

```python
#21. Calculate region wise average profit. 

Average_profit=ne.groupby('Region')['Profit'].mean()
Average_profit
```

```python
#22. Find the ship mode which is more profitable.

eee=ne.groupby('Ship_Mode')['Profit'].sum().idxmax()
eee
```

```python
#23. Create a new data frame where loss is recorded

Loss_recorded=ne[ne['Profit']<0]
Loss_recorded
```

```python
#24. Find category wise sub category wise sales

Category_wise_sale=ne.groupby(['Category','Sub-Category'])['Sales'].sum()
Category_wise_sale
```

```python
# 25. Create a data frame for record where discount is 0 

Discount_0=ne[ne['Discount']==0]
Discount_0
```
