```python
# 14. Find Records for June 2014

#ne['Order_Date']=pd.to_datetime(ne['Order_Date'])
june_2014=ne[(ne['Order_Date'] >= '2014-06-01') & (ne['Order_Date'] <= '2014-06-30')]
#june_2014=ne[(ne['Order_Date'].dt.month==6) & (ne['Order_Date'].dt.year==2014)]
june_2014



import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(june_2014['Ship_Date'], june_2014['Sales'], color='Yellow')
plt.title('Sales for June 2014')
plt.xlabel('Ship Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
#plt.show()

# print('Record of Jun Month')
# display(ne_june_month)
```

```python
# 15. Calculate month wise average of profit. 

ne['Order_Date']=pd.to_datetime(ne['Order_Date'])  
monthly_avg_profit=ne.groupby(ne['Order_Date'].dt.month)['Profit'].mean()
monthly_avg_profit

import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(monthly_avg_profit.index, monthly_avg_profit.values, color='Magenta')
plt.title('Profit Of Month')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
display(monthly_avg_profit)
```

```python
# 16. Create the new data frame for California 

california_data = ne[ne['State'] == 'California']
california_data
```

```python
# 17. Find sum of sales  

sum_of_sales = ne['Sales'].sum()
sum_of_sales

```

```python
#18 Find state wise sum of sales.

aaaa=ne.groupby('State')['Sales'].sum()
aaaa


plt.figure(figsize=(10, 7))
plt.bar(aaaa.index, aaaa.values, color='Magenta')
plt.title('Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
print("state wise sale\n", aaaa)
```

```python
#19. Find month wise sum  of sales 

bbb=ne.groupby([ne['Order_Date'].dt.year, ne['Order_Date'].dt.month])['Sales'].sum()
bbb

plt.figure(figsize=(10, 7))
plt.bar([str(i) for i in bbb.index], bbb.values, color='Yellow')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
print("Every Month Sale\n",bbb)
```

```python
#20. Find country wise and city wise average profit

ccc=ne.groupby(['Country','City'])['Profit'].mean()
ccc
```

```python
#21. Calculate region wise average profit. 

Average_profit=ne.groupby('Region')['Profit'].mean()
Average_profit

plt.figure(figsize=(8, 5))
plt.bar(Average_profit.index, Average_profit.values, color='skyblue')
plt.title('Average Profit by Region')
plt.xlabel('Region')
plt.ylabel('Average Profit')
plt.xticks(rotation=50)
plt.grid(True)
plt.show()
print("Average Profit by Region\n",Average_profit)
```

```python
#22. Find the ship mode which is more profitable.


most_profitable_ship_mode_profit = ne.groupby('Ship_Mode')['Profit'].sum().max()
most_profitable_ship_mode_data = ne.groupby('Ship_Mode')['Profit'].sum().reset_index()
plt.figure(figsize=(10, 7))
plt.bar(most_profitable_ship_mode_data['Ship_Mode'], most_profitable_ship_mode_data['Profit'], color='skyblue')
plt.title('Profit by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
